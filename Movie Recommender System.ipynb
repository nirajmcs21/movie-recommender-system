```python
#converting string to list
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres'] = movies['genres'].apply(convert);
```

```python
def convert1(obj):
    L1 = []
    for i in ast.literal_eval(obj):
        L1.append(i['name'])
    return L1
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if(counter !=3):
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['crew'].values
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
           
```

```python
movies.head()
```

```python
def splits(obj):
    return obj.split()
movies['overview'] = movies['overview'].apply(splits)
```

```python
# remove space in genres
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres']+ movies['keywords'] + movies['cast']+ movies['crew']
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
#converting into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

# Stemming is a natural language processing technique that lowers inflection in words to their root forms, hence aiding in the preprocessing of text, words, and documents for text normalization.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Removing similar words like ['love', 'loving', 'loved'] gives love
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# converting text into vectors


```python
#this process is called textvectorisation
# i used bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
#Now we have 4806 movies and we have 4806 vector as well, Now I calculate distance between all vectors
#Calculating angle between two vectors that is cosine distance (theta between two vectors)
#Cosine similarity always lies between 0 and 1, 1 means more similar, and 0 means opposite
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# i apply sorting in reverse order
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])
```

```python
# now i choose top 5
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    #fetch index
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
